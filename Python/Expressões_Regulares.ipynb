# Expressões regulares

. Qualquer caractere (exceto linha nova)
\w Quaquer caractere alfanumérico
\W Qualquer caractere não-alfanumérico
\d Qualquer caractere que seja um dígito (0-9)
\D Qualquer caractere que não seja dígito
\s Espaço em branco
^  Começa com 
$  Termina com
"\" usado antes de metacaracteres para especificar seu significado literal

[] Opcional entre os que estão dentro dos colchetes
() Captura grupos de caracteres
*  De zero a mais vezes
?  Zero ou uma vez
+  Uma ou mais vezes (obrigatório)
{m,n} De m a n vezes


## Função search

```python
import re
```

```python
frase = 'Olá, meu número de telefone é (42)00010-0000'
```

```python
#re.search('\(\d{2}\)\d{4,5}-\d{4}', frase)
#<>:1: SyntaxWarning: invalid escape sequence '\('
# Corrigido usando raw string
re.search(r'\(\d{2}\)\d{4,5}-\d{4}', frase)
```

```python
frase = 'A placa de carro que eu anotei durante o acidente foi FrT-1998'
```

```python
re.search(r'[A-Za-z]{3}-\d{4}', frase)
```

```python
email = 'Entre em contato, meu email é teste@teste.com'
```

```python
re.search(r'\w+@\w+\.com', email)
```

## Função match

```python
frase = 'A placa de carro que eu anotei durante a batida foi FRT-1998'
```

```python
print(re.match(r'[A-Za-z]{3}-\d{4}', frase))
```

```python
frase2 = 'FRT-1998 é a placa do carro'
```

```python
print(re.match(r'[A-Za-z]{3}-\d{4}', frase2))
```

## Função findall

```python
frase3 = 'Meu número de telefone atual é (42)0000-0000. O número (56)1111-1111 é o antigo'
```

```python
re.findall(r'\(\d{2}\)\d{4,5}-\d{4}', frase3)
```

```python
emails = '''Nome: Teste 1
email: teste1@teste.com
Nome: Teste 2
email: teste2@teste.com
Nome: Teste 3
email: teste3@teste.com.br
'''
```

```python
re.findall('\w+@\w+\.\w*', emails)
```

```python
re.findall('[\w.+-]+@[\w-]+\.[\w.-]+', emails) #Atualizado 08/02/2022
```

```python
re.findall('\w+@\w+\.\w+.?\w+', emails) #Colaboração/sugestão Eder Marinho Martins
```
